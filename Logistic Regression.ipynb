# Feature Engineering

```python
bank.y.replace(to_replace=['no','yes'],value=[0,1],inplace=True)
```

```python
bank['y'].value_counts()
```

```python
bank.columns
```

```python
bank=bank[['y','age', 'job', 'marital', 'education', 'default', 'balance', 'housing',
       'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome']]
```

```python
bank.columns
```

```python
pd_1=pd.get_dummies(bank,columns=[ 'job', 'marital', 'education', 'default', 'balance', 'housing',
       'loan', 'contact', 'month', 'campaign', 'pdays',
       'previous', 'poutcome'])
pd_1.head()
```

# Building Model

```python
#Dividing the data into input and output variables
X=pd_1.iloc[:,1:697]
Y=pd_1.iloc[:,0]
```

```python
Y
```

```python
X.head()
```

```python
Y.head()
```

# Fit The Model

```python
classifier= LogisticRegression()
classifier.fit(X,Y)
```

```python
classifier.intercept_
```

```python
np.round(classifier.coef_,2)
```

```python
#PREDICT for X dataset
y_pred = classifier.predict(X)
```

```python
y_pred_df= pd.DataFrame({'actual': Y,
                         'predicted_values': classifier.predict(X)})
```
